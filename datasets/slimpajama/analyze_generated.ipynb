```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("generated.csv")
df["char_length"] = df["text"].apply(lambda x: len(x))
df["word_length"] = df["text"].apply(lambda x: len(x.split()))
print(df["char_length"].describe())
print(df["word_length"].describe())
plt.hist(df["char_length"], bins=100)
plt.show()
plt.hist(df["word_length"], bins=100)
plt.show()
```
